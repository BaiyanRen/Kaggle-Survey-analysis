## Exploratory Data Analysis

### General features of DS and ML prectitioners with Doctoral degree

#### 11.49% of DS and ML practitioners have Doctoral degree!

Majority are Bachelor's and Master's degree.

```python
edu_count = survey_2020['Education'].value_counts()[Education]
total = survey_2020.shape[0]
```

```python
sns.set_style('dark')
fig, axes = plt.subplots(figsize=[12, 5])
edu_color = sns.cubehelix_palette(n_colors=7, start=2.8, rot=-0.1, gamma=0.9, hue=0.8, light=0.8, dark=0.3)
axes.bar(edu_count.index, edu_count, color=edu_color, width=0.4, edgecolor=(0, 0, 0))

for i in edu_count.index:
    axes.annotate('{:.2f}%'.format(edu_count[i]/total*100),
                  xy=(i, edu_count[i]+200),
                  va='center', 
                  ha='center', 
                  alpha=0.7)

plt.yticks(fontsize=12)
plt.ylabel('Number of Participants', fontsize=12)

plt.xticks(rotation=90, fontsize=12)

axes.axhline(color='cadetblue', lw=4)

axes.set_title('Distribution of Education Level', fontsize=16, fontweight='bold', loc='left');
```

#### Gender distribution
Males comprise the majority in all education levels. The ratio of male to female in Doctoral degree (3.62) is smaller than that in all education levels (4.07).

```python
gender = ['Man', 'Woman', 'Prefer not to say', 'Nonbinary', 'Prefer to self-describe']

gender_total = survey_2020['Gender'].value_counts()[gender]
gender_prop_total = -gender_total/gender_total.sum()*100

gender_doct = survey_2020_doct['Gender'].value_counts()[gender]
gender_prop_doct = gender_doct/gender_doct.sum()*100
```

```python
fig, ax = plt.subplots(figsize=[15, 9])

ax.bar(gender_doct.index, height = gender_prop_doct, color=edu_color[3], edgecolor=(0, 0, 0), width=0.3, label='Doctoral degree')
ax.bar(gender_total.index, height= gender_prop_total, color='silver', edgecolor=(0, 0, 0), width=0.3, label='All')

for i in gender_total.index:
    ax.annotate('{:.2f}%'.format(-gender_prop_total[i]), 
                xy=(i, gender_prop_total[i]-4), 
                va='center', 
                ha='center')
    
for j in gender_doct.index:
    ax.annotate('{:.2f}%'.format(gender_prop_doct[j]),
                xy=(j, gender_prop_doct[j]+3),
                va='center',
                ha='center')

plt.grid(which='major', axis='x')

yticks = [-80, -60, -40, -20, 0, 20, 40, 60, 80]   
plt.yticks(yticks, np.abs(yticks), fontsize=12)
plt.ylabel('Percentage', fontsize=12)

plt.xticks(fontsize=12)

plt.title('Distribution of Gender in Doctoral Degree Level (%)', fontsize=16, fontweight='bold', loc='left')

plt.legend(['Doctoral degree', 'All'], title='Education levels');
```

```python
print('The male to female ratio in all education levels is {:.2f}.'.format(gender_total['Man']/gender_total['Woman']))
```

```python
print('The male to female ratio in all education levels is {:.2f}.'.format(gender_doct['Man']/gender_doct['Woman']))
```

#### Age distribution

```python
age_total = survey_2020.groupby('Age').size()
age_prop_total = -age_total/age_total.sum()*100
age_doct = survey_2020_doct.groupby('Age').size()
age_prop_doct = age_doct/age_doct.sum()*100
```

```python
fig, ax = plt.subplots(figsize=[15, 9])

ax.bar(age_doct.index, age_prop_doct, color=edu_color[4], edgecolor=(0, 0, 0), width=0.6, label='Doctoral degree')
ax.bar(age_total.index, age_prop_total, color='silver', edgecolor=(0, 0, 0), width=0.6, label='All')

for j in age_doct.index:
    ax.annotate('{:.2f}%'.format(age_prop_doct[j]),
                xy=(j, age_prop_doct[j]+0.8),
                va='center', 
                ha='center')
for i in age_total.index:
    ax.annotate('{:.2f}%'.format(-age_prop_total[i]),
                xy=(i, age_prop_total[i]-1),
                va='center',
                ha='center')

plt.grid(which='major', axis='x')

yticks=[-20, -10, 0, 10, 20]
plt.yticks(yticks, np.abs(yticks), fontsize=12)
plt.ylabel('Percentage', fontsize=12)

plt.xticks(fontsize=12)

plt.title('Distribution of Age in Doctoral Degree and All Education Levels (%)', 
          fontsize=16, 
          fontweight='bold',
          loc='left')
plt.legend(['Doctoral degree', 'All'], title='Education levels');
```

For survey participants with Doctoral degree, majority are older than 30 years old. The peak of distribution appears in age range 30-34 years old. 

For all the education levels, majority of people are younger than 30 years old. The peak of the age distribution locates in age range 25-29 years old. 

Considering the long period of time to complete the Doctoral degree (at least 5 years), it is not surprising to observe the right-switch of the age distribution in Doctoral degree.

Given the different economical condition in different countries, I shrink the range into USA and India, which are the top two countries from which the most participants came. In addition, they are two representative developed and developing countries.

```python
def title_vs_salary(country):
    survey_2020_c = survey_2020[survey_2020['Country']==country].copy()
    title_salary_c = survey_2020_c.groupby(['Title', 'Salary']).size().unstack()[salary].fillna(0).astype(int)
    
    fig, ax = plt.subplots(figsize=[15, 9])

    sns.heatmap(data=title_salary_c,
                  cmap='bone_r', 
                  linewidths=0.2, 
                  square=True, 
                  annot=True, 
                  fmt = 'd', 
                  annot_kws={'alpha': 0.9},
                  cbar_kws={'shrink': 0.6, 'label': 'Number of participants'})

    plt.title('Distribution of Title and Salary in {}'.format(country), fontsize=16, fontweight='bold', loc='left', va='bottom');
```

```python
title_vs_salary('United States of America')
```

```python
title_vs_salary('India')
```

There is a striking difference in salary distribution between USA and India.
